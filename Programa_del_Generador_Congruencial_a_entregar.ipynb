**Programa del Generador Congruencial**

---



---


Nuestro programa solicitará al ususario la constante "a", la constante "c" y la constante "m". De igual manera solicitará un número al que llamaremos "semilla x_n", de modo que nos da:

x_n+1 = a * x_n + c

Hay que tomar en cuenta que nuestra "semilla" debe ser mayor a "0" y menor a nuestra constante "m"

Nuestra tabla se mostrará de la siguiente manera:


*   Partiendo de nuestras "constantes" y nuestra semilla" se obtendrá un resultado el cuál se mostrará en la columna "producto"
*   Nuestra columna "sobrante" son los enteros que sobran al dividir el "producto" sobre la constante "m"
*   Por último, pero no menos importante. La columna "resultado" es el valor obtenido al dividir la "semilla" sobre la constante "m"

El único valor alterado será nuestra "semilla" (nuestras constantes permanecen iguales). Por lo que ahora el ciclo partirá considerando las mismas "constantes", pero una "semilla" distinta.

---



---

El programa pedirá al ususario que una constante "a", "c" y m".
De igual manera una "semilla", la cuál debe ser mayor que "0" y menor que "m", en caso contrario el programa pedirá otra "semilla"

```python
#Importamos "tabulate" para hacer nuestra tabla de una manera más presentable
from tabulate import tabulate

print("******************************************")
print("* Programa del Generador Congruencial *")
print("******************************************\n")

#Solicitamos al las constantes necesarias para ejecutar el programa
num_alta = int(input("Ingresa un número el cúal será tu constante 'a': "))
num_altc = int(input("Ingresa un número el cúal será tu constante 'c': "))
num_altm = int(input("Ingresa un número el cúal será tu constante 'm': "))

while True:
    num_altxn = int(input("Ingresa un número el cúal será tu semilla 'x_n': "))
    if 0 < num_altxn < num_altm:
        break
    else:
        print("Por favor, ingresa una semilla válida. Debe ser mayor a '0' y menor a 'm'.")
```

Ahora el programá pedirá las iteraciones que deseemos realizar.

```python
#Solicitamos al usuario la cantidad de iteraciones
num_iteraciones = int(input("Ingresa la cantidad de iteraciones que deseas realizar: "))
```

Todo se almacenará en un lista, que estará reflejado en nuestra tabla.

```python
#Almacenamos los resultados en una lista
resultados = []

#Construimos los datos para la tabla
for i in range(num_iteraciones):
    num_alt2 = num_altxn * num_alta + num_altc
    num_alt3 = num_alt2 % num_altm
    num_alt4 = num_altxn / num_altm

    #Almacenamos los resultados en la lista
    resultados.append((i, num_altxn, num_alt2, num_alt3, num_alt4))

    #Establecemos el nuevo valor de num_altxn para la siguiente iteración
    num_altxn = num_alt3

#Imprimimos la tabla
headers = ["Iteración", "Semilla", "Producto", "Sobrante", "Resultado"]
print(tabulate(resultados, headers=headers, tablefmt="grid"))
```
