```python
from tabulate import tabulate

print("********************************")
print("* Programa del Cuadrado Medio *")
print("********************************\n")

# Solicitamos al usuario un número de 4 dígitos
while True:
    num_alt = input("Ingresa un número de 4 dígitos: ")
    if num_alt.isdigit() and len(num_alt) == 4:
        num_alt = int(num_alt)
        break
    else:
        print("Por favor, ingresa un número válido de 4 dígitos.")

# Solicitmos al usuario la cantidad de iteraciones
num_iteraciones = int(input("Ingresa la cantidad de iteraciones que deseas realizar: "))

# Almacenamos los resultados en una lista
resultados = []

# Construimos los datos para la tabla
for i in range(num_iteraciones):
    num_alt2 = num_alt * num_alt

# Convertimos el resultado a cadena y obtener la longitud
    result_str = str(num_alt2)
    result_length = len(result_str)

#Por ejemplo:
#Calculamos el cuadrado de "1234", que es "1522756".
#Luego, calculamos la longitud de la cadena resultante, que es result_long = len("1522756"), por lo que result_long es 7.
#Ahora, calculamos la posición inicial para los 4 dígitos del medio: ini_posicion = (7 - 4) // 2, que es igual a 1.
#La posición final sería: fin_posicion = ini_posicion + 4, por lo que fin_posicion sería 5.
# Calculamos la posición inicial y final para los 4 dígitos de en medio
    inicio = (result_length - 4) // 2
    final = inicio + 4

    # Extraemos los 4 dígitos de en medio
    medio = result_str[inicio:final]

    # Almacenamos los resultados en la lista
    resultados.append((i+1, num_alt, num_alt2, medio))

    # Establecemos el nuevo valor de num_alt para la siguiente iteración con la codición de agregar un "0" a los números con dígitos impares
    num_alt = int("0" + medio) if result_length % 2 != 0 else int(medio)

# Imprimir la tabla
headers = ["Iteración", "Cuadrado medio", "Cuadrado medio al cuadrado", "Cuadrado medio continuo"]
print(tabulate(resultados, headers=headers, tablefmt="grid"))
```
