# Estimación del número 'pi'

Importamos la librarías necesarias

```python
import numpy as np
import matplotlib.pyplot as plt
```

Gráficamos el la estimación.

```python
# Definimos el número de puntos N
N = 100

# Generamos dos arreglos de números aleatorios entre 0 y 1
u1 = np.random.rand(N)
u2 = np.random.rand(N)

# Inicializamos un contador para los puntos dentro del cuarto de círculo
cont = 0

# Lista para almacenar las estimaciones de π
pi_estimations = []

# Recorremos cada punto
for i in range(N):
    # Verificamos si el punto (u1[i], u2[i]) está dentro del cuarto de círculo
    if u2[i] <= np.sqrt(1 - u1[i]**2):
        cont += 1
        # Si está dentro, lo graficamos en color rojo
        plt.scatter(u1[i], u2[i], color="red", s=1)
    else:
        # Si está fuera, lo graficamos en color azul
        plt.scatter(u1[i], u2[i], color="blue", s=1)

    # Calculamos la estimación de π y la almacenamos
    pi_estimations.append((cont / (i + 1)) * 4)

# Graficamos la curva del cuarto de círculo
x = np.linspace(0, 1, 1000)
y = np.sqrt(1 - x**2)
plt.plot(x, y, label='y = sqrt(1 - x^2)', color='black')

# Añadimos etiquetas y título a la gráfica
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Estimación de π ≈ {pi_estimations[-1]}')
plt.legend()
plt.show()
```

Guardamos los resultados para simular la parte transitoria y la estacionaria

```python
# Imprimimos la estimación final de π
print(f'Estimación de π: {pi_estimations[-1]}')

# Graficamos las estimaciones de π para ver la parte transitoria y la estacionaria
plt.figure(figsize=(10, 6))
plt.plot(pi_estimations, label='Estimación de π')
plt.axhline(y=np.pi, color='r', linestyle='-', label='Valor verdadero de π')
plt.xlabel('Número de iteraciones')
plt.ylabel('Estimación de π')
plt.title('Convergencia de la estimación de π')
plt.legend()
plt.grid(True)
plt.show()
```
