```python
#Importamos tabulate de esta manera
from tabulate import tabulate

print("******************************************************************************************")
print("Este programa toma 4 dígitos de un número, lo eleva al cuadrado y toma el cuadrado medio")
print("****************************************************************************************** \n")

# Solicitar al usuario un número de 4 dígitos
while True:
    num_alt = input("Ingresa un número de 4 dígitos: ")
    if num_alt.isdigit() and len(num_alt) == 4:
        num_alt = int(num_alt)
        break
    else:
        print("Por favor, ingresa un número válido de 4 dígitos.")

# Solicitar al usuario la cantidad de iteraciones
num_iteraciones = int(input("Ingresa la cantidad de iteraciones que deseas realizar: "))

# Se almacenan los resultados en una lista
resultados = []

#Damos la indicación de cuántas iteraciones deseamos hacer
for _ in range(num_iteraciones):
    num_alt2 = num_alt * num_alt

#De la siguiente manera agregamos un '0' a nuestro número elevado al cuadrado, en caso de tener menos de 7 dígitos
    cadena = str(num_alt2)
    cantidad = len(cadena)
    cero = str(0)
    if cantidad < 8:
        num_alt3 = cero + cadena
        num_alt4 = [num_alt3[i] for i in range(1, 5)]
        resultados.append(num_alt4)
    else:
        cadena_alt = str(num_alt2)
        num_alt2 = [cadena_alt[i] for i in range(2, 6)]
        resultados.append(num_alt2)

# Imprime los resultados en forma de tabla
headers = ["Iteración", "Cuadrado Medio", "Cuadrado Medio Continuo"]
tabla_resultados = [[i + 1, ''.join(map(str, r)), int(''.join(map(str, r))) ] for i, r in enumerate(resultados)]

print(tabulate(tabla_resultados, headers=headers, tablefmt="fancy_grid"))
```
