# Experimento de la aguja de Buffon

En esta simulación se hizó la simulación de arrojar 1000 veces las agujas.
---
---
---



Importamos la librarías necesarias

```python
import numpy as np
import matplotlib.pyplot as plt
```

Gráficamos el la estimación.

```python
# Parámetros del experimento
L = 1  # Longitud de la aguja
d = 2  # Distancia entre las líneas
N = 1000  # Número de lanzamientos

# Inicialización del contador para las agujas que cruzan una línea
# Se cambio el nombre del contador, el cual normalmente es 'cont'
cruces = 0

# Lista para almacenar las estimaciones de cruces
cruces_estimacion = []

# Realización del experimento
for i in range(N):
    # Generar un ángulo aleatorio entre 0 y π
    theta = np.random.uniform(0, np.pi)
    # Generar una distancia aleatoria desde el centro de la aguja a la línea más cercana
    x = np.random.uniform(0, d / 2)

    # Verificar si la aguja cruza una línea
    if x <= (L / 2) * np.sin(theta):
        cruces += 1
        # Si cruza, lo graficamos en color rojo
        plt.scatter(x, theta, color="red", s=1)
    else:
        # Si no cruza, lo graficamos en color azul
        plt.scatter(x, theta, color="blue", s=1)

    # Almacenamos la estimación del cruce
    cruces_estimacion.append(cruces / (i + 1))

# Graficamos los resultados de la simulación
plt.xlabel('Distancia desde el centro de la aguja a la línea más cercana')
plt.ylabel('Ángulo de la aguja')
plt.title('Simulación del experimento de la aguja de Buffon')
plt.show()
```

Guardamos los resultados para simular la parte transitoria y la estacionaria

```python
# Graficamos la probabilidad de cruce para ver la parte transitoria y la estacionaria
plt.figure(figsize=(10, 6))
plt.plot(cruces_estimacion, label='Probabilidad de Cruce')
plt.xlabel('Número de lanzamientos')
plt.ylabel('Probabilidad de Cruce')
plt.title('Convergencia de la Probabilidad de Cruce en el Experimento de la Aguja de Buffon')
plt.legend()
plt.grid(True)
plt.show()
```
