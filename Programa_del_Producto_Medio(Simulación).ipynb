```python
from tabulate import tabulate

print("********************************")
print("* Programa del Producto Medio *")
print("********************************\n")

    # Solicitamos al usuario el primer número de 3 dígitos
while True:
    num_alt = input("Ingresa un número de 3 dígitos como primer factor: ")
    if num_alt.isdigit() and len(num_alt) == 3:
        num_alt = int(num_alt)
        break
    else:
        print("Por favor, ingresa un número válido de 3 dígitos.")

    # Solicitamos al usuario el segundo número de 3 dígitos
while True:
    num_alt2 = input("Ingresa un número de 3 dígitos como segundo factor: ")
    if num_alt2.isdigit() and len(num_alt2) == 3:
        num_alt2 = int(num_alt2)
        break
    else:
        print("Por favor, ingresa un número válido de 3 dígitos.")

    # Solicitamos al usuario la cantidad de iteraciones
num_iteraciones = int(input("Ingresa la cantidad de iteraciones que deseas realizar: "))

    # Almacenamos los resultados en una lista
resultados = []

    # Construimos los datos para la tabla
for i in range(num_iteraciones):
    # Multiplicamos los dos números
    num_alt3 = num_alt * num_alt2

    # Convertimos el resultado a cadena y obtenemos la longitud
    result_str = str(num_alt3)
    result_length = len(result_str)

    # Verificamos si la longitud es par, y si es así, añadimos un '0' al principio
    # result_length += 1. Se ha agregado un '0' al principio, por lo que la longitud debe incrementarse.
    if result_length % 2 == 0:
        result_str = '0' + result_str
        result_length += 1

    # Calculamos la posición inicial y final para los 3 dígitos de en medio
    inicio = (result_length - 3) // 2
    final = inicio + 3

    # Extraemos los 3 dígitos de en medio
    medio = result_str[inicio:final]

    # Almacenar los resultados en la lista
    resultados.append((i + 1, num_alt, num_alt2, num_alt3, medio))

    # Establecemos el nuevo valor de num_alt y num_alt2 para la siguiente iteración
    num_alt = num_alt2
    num_alt2 = int(medio)

    # Imprimimos la tabla
headers = ["Iteración", "Primer Factor", "Segundo Factor", "Producto", "Dígitos de en medio"]
print(tabulate(resultados, headers=headers, tablefmt="grid"))
```
